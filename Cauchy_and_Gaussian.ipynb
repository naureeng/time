```python
import numpy as np
import matplotlib.pyplot as plt
N = 100000

G1_x = np.random.randn(N)
G1_y = np.random.randn(N)

plt.scatter(G1_x, G1_y, marker='o', c='r')
plt.show()
```

```python
N = 100000

G = np.random.randn(N)
a1 = np.random.randn(N)
G_x = G*np.sin(np.pi*2*a1)
G_y = G*np.cos(np.pi*2*a1)

plt.scatter(G_x, G_y, marker='o', c='b')
plt.show()
```

```python
C = np.random.standard_cauchy(N)
a1 = np.random.rand(N)
C_x = C*np.sin(np.pi*2*a1)
C_y = C*np.cos(np.pi*2*a1)

C1_x = np.random.standard_cauchy(N)
C1_y = np.random.standard_cauchy(N)

plt.scatter(C_x, C_y, marker='o', c='b')
plt.scatter(C1_x, C1_y, marker='o', c='r')
plt.xlim([-500,500])
plt.ylim([-500,500])
plt.show()
```

```python
G = np.random.randn(N)
G1_x = np.random.randn(N)
G1_y = np.random.randn(N)

plt.scatter(G1_x, G1_y, marker='o', c='r')

C = 0.01*np.random.standard_cauchy(N)
a1 = np.random.randn(N)
C_x = C*np.sin(np.pi*2*a1)
C_y = C*np.cos(np.pi*2*a1)

plt.scatter(C_x, C_y, marker='o', c='b')
plt.ylim([-5,5])
plt.xlim([-5,5])
plt.show()

```
