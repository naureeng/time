# Ahrens and Sahani, 2008

Code to reproduce figures for paper 

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
from scipy import optimize

%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('font', family='Arial')
import seaborn as sns
```

```python
n = 100
xs = np.linspace(0, 20, num=n)
D = squareform(pdist(xs[:, np.newaxis]))
mn = np.zeros_like(xs)

# exponential kernel
K = np.exp(-0.5 * D)
data = np.random.multivariate_normal(mn, K)
for i in range(5):
    plt.plot(xs, np.random.multivariate_normal(mn, K))
    plt.rcParams['xtick.labelsize'] = 20
    plt.rcParams['ytick.labelsize'] = 20
    plt.xlabel('Time [s]', fontsize=20)
    plt.ylabel('Voltage [mV]', fontsize=20)
    sns.despine(trim=True, offset=1)
```

```python
sampling_rate = 30.0
fourier_transform = np.fft.rfft(data)
abs_fourier_transform = np.abs(fourier_transform)
power_spectrum = np.square(abs_fourier_transform)
frequency = np.linspace(0, sampling_rate/2, len(power_spectrum))
plt.loglog(frequency, power_spectrum)
plt.xlabel('Temporal Frequency [Hz]', fontsize=20)
plt.ylabel('log [Power] [W]', fontsize=20)
sns.despine(trim=True, offset=1)
```
